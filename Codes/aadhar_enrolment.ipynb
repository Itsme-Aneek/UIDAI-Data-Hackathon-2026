```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Convert age columns into single column
age_state = df.melt(
    id_vars='state',
    value_vars=['age_0_5', 'age_5_17', 'age_18_greater'],
    var_name='age_group',
    value_name='count'
)

# Step 2: Top 5 states by total enrolment
top_states = (
    age_state.groupby('state')['count']
    .sum()
    .sort_values(ascending=False)
    .head(5)
    .index
)

age_state = age_state[age_state['state'].isin(top_states)]

# Step 3: Plot
plt.figure(figsize=(10,6))
sns.barplot(
    data=age_state,
    x='state',
    y='count',
    hue='age_group'
)

plt.title("Top 5 States: Age-wise Aadhaar Enrolment Comparison")
plt.xticks(rotation=45)
plt.xlabel("State")
plt.ylabel("Enrolment Count")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Convert date column to datetime
df['date'] = pd.to_datetime(df['date'], format='%d-%m-%Y')

# Step 2: Extract month
df['month'] = df['date'].dt.to_period('M').astype(str)

# Step 3: Create total enrolment column
df['total_enrolment'] = (
    df['age_0_5'] +
    df['age_5_17'] +
    df['age_18_greater']
)

# Step 4: Group by month and state
trend_state = (
    df
    .groupby(['month', 'state'])['total_enrolment']
    .sum()
    .reset_index()
)

# Step 5: Top 3 states by overall enrolment
top_states = (
    df
    .groupby('state')['total_enrolment']
    .sum()
    .sort_values(ascending=False)
    .head(3)
    .index
)

trend_state = trend_state[trend_state['state'].isin(top_states)]

# Step 6: Plot
plt.figure(figsize=(12,6))
sns.lineplot(
    data=trend_state,
    x='month',
    y='total_enrolment',
    hue='state',
    marker='o'
)

plt.title("Monthly Aadhaar Enrolment Trend for Top 3 States")
plt.xlabel("Month")
plt.ylabel("Total Enrolment")
plt.xticks(rotation=45)
plt.show()
```

```python
daily = df.groupby('date_dt').size().reset_index(name='count')

threshold = daily['count'].mean() + 3*daily['count'].std()
anomalies = daily[daily['count'] > threshold]
anomalies
```

```python
plt.figure(figsize=(12,5))
plt.plot(daily['date_dt'], daily['count'])
plt.scatter(anomalies['date_dt'], anomalies['count'], color='red')
plt.title("Anomalies in Aadhaar Enrolment Volume")
plt.show()
```
